```python
import numpy as np

x1=[0,0,1,1]
x2=[0,1,0,1]
OP=[0,1,1,0]
w1=0.45
w2=0.56
b=0.33
number_of_epochs=1500
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

for epoch in range(1,number_of_epochs+1):
    print(f'Epoch Number:{epoch}')
    for iter in range(len(OP)):
        z=w1*x1[iter]+w2*x2[iter]
        print(f'calculated z:{z}')
        P=sigmoid(z)
        print(f'predicted Value:{P}')
        cost_val=(P-OP[iter])**2
        print(f'cost value: {cost_val}')
        #print(f'iter number:{iter}')
    print('--'*25)
```

```python
import numpy as np
h=0.0001
alpha=0.01
def cost(w1,w2,b,A,x1,x2):
    z=w1*x1+w2*x2+b
    p=1/(1+np.exp(-z))
    return (p-A)**2

for epoch in range(1,number_of_epochs+1):
    print(f'Epoch number:{epoch}')
    for iter in range(len(OP)):
        d_J_w1=cost(w1,w2,b,OP[iter],x1[iter],x2[iter])
        d_j_w1_h=cost(w1+h,w2,b,OP[iter],x1[iter],x2[iter])
        d_w1=(d_J_w1-d_j_w1_h)/h
    print('---'*25)

    for iter in range(len(OP)):
        d_J_w2=cost(w1,w2,b,OP[iter],x1[iter],x2[iter])
        d_j_w2_h=cost(w1,w2+h,b,OP[iter],x1[iter],x2[iter])
        d_w2=(d_J_w2-d_j_w2_h)/h
    print('---'*25)

    for iter in range(len(OP)):
        d_J_b=cost(w1,w2,b,OP[iter],x1[iter],x2[iter])
        d_j_b_h=cost(w1,w2,b+h,OP[iter],x1[iter],x2[iter])
        d_b=(d_J_b-d_j_b_h)/h
    print('---'*25)
    w1=w1-alpha*d_w1
    w2=w2-alpha*d_w2
    b=b-alpha*d_b

    #print(w1,w2,b)

sigmoid(w1*1+w2*0+b)



```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers,models

model=models.Sequential()
model.add(layers.Dense(64,activation='relu', input_dim=2))
model.add(layers.Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='sgd',metrics=['accuracy'])
```

```python
import pandas as pd

data=pd.DataFrame(columns=['x1','x2','OP'])
data['x1']=x1
data['x2']=x2
data['OP']=OP

X=data.drop(columns='OP').values
print(X)
y=np.array(data['OP']).reshape(-1,1)
print(y)

model.fit(X,y,epochs=300)
```
