```python
import numpy as np
import random

# Define your matrix A (e.g., [[2, 1], [1, 2]])
# Initialize x (random initial vector)
#A=np.array([[2,1],[1,2]])

#print(x)

def get_matrix_from_user():
    # Prompt the user for matrix dimensions
    rows = int(input("Enter the number of rows: "))
    cols = int(input("Enter the number of columns: "))

    # Initialize an empty list to store the matrix elements
    matrix = []

    print("Enter the matrix elements row by row (separated by space):")
    for i in range(rows):
        while True:
            # Get the row elements from the user
            row_input = input(f"Enter elements of row {i+1}: ")
            row = list(map(float, row_input.split()))
            
            # Check if the correct number of elements were entered
            if len(row) != cols:
                print(f"Error: Please enter exactly {cols} elements.")
            else:
                matrix.append(row)
                break

    # Convert the list to a NumPy array and return
    return np.array(matrix)


def main():
# Set tolerance and maximum iterations
    A=get_matrix_from_user()
    print(A)
    #x = np.random.rand(A.shape[0])
    x1=[[0],[1],[0]]
    x=np.array(x1)
    print(x.shape)
    tol = 1e-6
    max_iter = 100
    lambda_old = 0  # Initialize previous eigenvalue
    for _ in range(max_iter):
        x = np.matmul(A, x)
        print(x)
        x /= np.linalg.norm(x)
        print(x)
        lambda_new = max(abs(x))
        print(lambda_new)

        if abs(lambda_new - lambda_old) < tol:
         break

        lambda_old = lambda_new

# Final eigenvalue and eigenvector
    largest_eigenvalue = lambda_new
    eigenvector = x
    print("Largest Eigenvalue:", largest_eigenvalue)
    print("Eigenvector:", eigenvector)

if __name__ == "__main__":
    main()
```
